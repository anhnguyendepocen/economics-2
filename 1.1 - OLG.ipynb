### 1.1 - OLG

```python
"""OLG example code: Python in a procedural style"""
import matplotlib.pyplot as plt

def simulate(alpha=0.3, k0=0.01):
  # Set parameter and initial conditions
  kss = ((1 - alpha) / 2)**((1 / (1 - alpha)));

  # Set stopping criteria / max iterations
  Tmax    = 100;
  epsilon = 0.001;
  delta   = 0.001;

  # Setup model variables
  k = [k0];

  # Iterate
  while True:
    k.append(
      ((1 - alpha) / 2) * (k[-1]**alpha)
    );

    # Break out of the iteration if we are close to equilibrium, or aren't moving
    if abs(k[-1] - kss) < epsilon or abs(k[-1] - k[-2]) < delta:
      break;
    # Break out of the iteration if we've exceeded max
    if len(k) > Tmax-1:
      print 'Stopping criteria not met!';
      break;

  return (alpha, kss, k);
    
# Run the simulation
alpha, kss, k = simulate();

# Plot the results
T = len(k);
fig   = plt.figure();
axes  = fig.add_subplot(1,1,1);

# plot capital path
line, = axes.plot(range(T), k, color='black'); # note the tuple unpacking here

# add a stem plot
marker, stems, base = axes.stem(range(T), k, ':', bottom=kss);

# and a line showing equilibrium capital
axes.plot(range(T), [kss]*T, color='gray');

# prettify axes and lables
axes.set_title(r'OLG Capital Path, $\alpha = %s$' % alpha);
axes.set_xlabel('Time');
axes.set_ylabel('Capital');
axes.set_ylim(0, kss+0.2);

# prettify colors
map(lambda stem: stem.set(color='gray'), stems);
marker.set(color='gray');
```
