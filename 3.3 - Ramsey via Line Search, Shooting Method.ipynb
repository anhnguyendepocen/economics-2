### Example 3.3 - Ramsey via Line Search / Shooting Method

#### Ramsey Model

The first order conditions give us:

(Euler equation) $u'(c_t) = \beta(1 + f'(k_{t+1}) - \delta))u'(c_{t+1})$

(Capital accumulation equation) $k_{t+1} = f(k_t) + (1 - \delta)k_t - c_t$

And we assume that $k_0$ is given.

#### Functional Forms

We assume a CES utility function and a Cobb-Douglas production function.

$$
u(c_t) = \frac{(c^{1-\sigma} - 1)}{(1-\sigma)} ~\\\
f(k_t) = k_t^\alpha
$$

$$
u'(c_t) = c^{-\sigma} ~\\\
f'(k_t) = \alpha k_t^{\alpha-1}
$$

The parameters are set to be:

$$
\alpha = 0.5, ~ \beta = 0.95, ~ \delta = 0.1, ~ \sigma = 1.01
$$

So the final functional forms of the derivatives are:

$$
u'(c_t) = c_t^{-1.01} = \frac{1}{c_t^{1.01}} ~\\\
f'(k_t) = 0.5 k_t^{-0.5} = \frac{1}{2k_t^{0.5}} ~\\\
$$

#### Problem Setup
The shooting method specifies a terminal time period $T$ and a terminal level of capital $k_T$. (Also, we guess a $c_0$). Then we have to solve the following system of equations:

($T$ equations defining the evolution of $c$)

$$u'(c_0) = \beta(1 + f'(k_{1}) - \delta)u'(c_{1})$$
$$u'(c_1) = \beta(1 + f'(k_{2}) - \delta)u'(c_{2})$$
$$\vdots$$
$$u'(c_{T-1}) = \beta(1 + f'(k_{T}) - \delta)u'(c_{T})$$

(and another $T$ equations defining the evolution of $k$)

$$k_{1} = f(k_0) + (1 - \delta)k_0 - c_0$$
$$k_{2} = f(k_1) + (1 - \delta)k_1 - c_1$$
$$\vdots$$
$$k_{T} = f(k_{T-1}) + (1 - \delta)k_{T-1} - c_{T-1}$$

#### Solution Strategy: Line Search

Our objective is to solve $F(x) = 0~~$, where $F: \mathbb{R}^{2T} \to \mathbb{R}^{2T}$ is given by:

$$
F(c_0, \cdots, c_T, k_1, \cdots, k_{T-1}) = \begin{bmatrix}
  u'(c_0) - \beta(1 + f'(k_{1}) - \delta)u'(c_{1}) ~~~\\\
  \vdots ~~~\\\
  u'(c_{T-1}) - \beta(1 + f'(k_{T}) - \delta)u'(c_{T}) ~~~\\\
  k_{1} - f(k_0) - (1 - \delta)k_0 + c_0 ~~~\\\
  \vdots ~~~\\\
  k_{T} - f(k_{T-1}) - (1 - \delta)k_{T-1} + c_{T-1} ~~~\\\
\end{bmatrix}
$$

To do that, we reinterpret the problem as an optimization problem:

$$\max_x ~ G(x)$$

where $G: \mathbb{R}^{2T} \to \mathbb{R}$ is given by $G(x) = -F(x)'F(x)$

We implement this maximization via a "hill climbing" algorithm, where we follow these steps:

1. Pick $x0$
2. Pick the search direction $v_n$ (we will be using the direction implied by Newton's Method): $$v_n = -D^2G^{-1}(x_n) DG(x_n)~\\\$$
3. Pick the step size: $$\alpha_{n+1} = \text{argmax }_{\alpha \in \mathbb{R}} G(x_n + \alpha v_n)~\\\$$
4. Iterate: $$x_{n+1} = x_n + \alpha_{n+1} v_n ~\\\$$
5. Check stopping criteria and repeat


#### Shooting Method

Ultimately, we will form a guess of the entire consumption and capital vectors.

The consumption vector of interest is $\\{c_t\\}_{t=0}^{T}$

The capital vector that we _guess_ is $\\{k_t\\}_{t=1}^{T-1}$, because $k_0$ and $k_T$ are both specified (and we end up prepending $k_0$ and appending $k_T$ on each iteration to the $\\{k_t\\}_{t=1}^{T-1}$ guess).

#### Initialization

We will specify the following values:

$$
k_0 = k_{ss} - 1 ~\\\
k_T = k_{ss} - 0.01 ~\\\
T = 50 ~\\\
$$

Our initial guess will be $\\{c_t\\}_{t=0}^{T} = 0.5$ and $\\{k_t\\}_{t=1}^{T-1} = 8$

```python
import numpy as np
import sympy as sp
from sympy.utilities.lambdify import lambdify
from scipy.optimize import minimize


# Model parameters
#alpha,beta,delta,sigma = sp.symbols('alpha beta delta sigma')
a = 0.5;  # capital's share of income
b = 0.95; # discount rate
d = 0.1;  # depreciation rate
s = 1.01; # coefficient of relative risk aversion

# Steady stata values
#css,kss = sp.symbols('c_ss k_ss')
kss = ((1/b + d - 1)/a)**(1/(a - 1));
css = kss**a - d*kss;

# Functional specifications
ct,ct1,kt,kt1 = sp.symbols('c_t c_t+1 k_t k_t+1')
class u(sp.Function):
    nargs = 1
    @classmethod
    def eval(self, c):
        return (c**(1-s) - 1) / (1 - s)
    def fdiff(self, argindex = 1):
        return (self.args[0])**(-s)
class f(sp.Function):
    nargs = 1
    @classmethod
    def eval(self, k):
        return k**a
    def fdiff(self, argindex = 1):
        return a*(self.args[0]**(a-1))

# Initial and terminal conditions
#k0, kT, T = sp.symbols('k_0 k_T T')
k0 = kss - 7
kT = kss - 0.01
T  = 50
```

```python
# Create the system of equations F (i.e. we're trying to find F(x) = 0)
cn = sp.symbols('c:'+str(T+1))
kn = sp.symbols('k:'+str(T+1))
Du = [u(cn[i]).diff() for i in range(T+1)] # Cache derivatives for c0 - cT-1
Df = [f(kn[i]).diff() for i in range(T)] # Cache derivatives for k1 - kT-1
cc = [Du[i] - b*(1 + Df[i+1] - d)*Du[i+1] for i in range(T-1)] # T-1 functions
kk = [-(kn[i+1] - f(kn[i]) - (1-d)*kn[i] + cn[i]) for i in range(1,T-1)] # T-2 functions
# a few functions have some known values that we want to substitute in
cc = cc + [Du[T-1] - b*(1 + Df[0].subs({kn[0]:kT}) - d)*Du[T]] # Dfs[0].subs({kn[0]:kT}) is just f'(kT)
kk = [-(kn[1] - f(k0) - (1-d)*k0 + cn[0])] + kk                    # substitute in k0
kk = kk + [-(kT - f(kn[T-1]) - (1-d)*kn[T-1] + cn[T-1])]                 # substitute in kT

F = sp.Matrix(cc+kk)
args = cn + kn[1:T]

# Objective, Jacobian, and Hessian
G = -(F.dot(F))
J = sp.Matrix([G]).jacobian(args) # a gradient, really
H = J.jacobian(args)

# Get functions we can call
_F = lambdify(args, F, modules='numpy')
_G = lambdify(args, G, modules='numpy')
_J = lambdify(args, J, modules='numpy')
_H = lambdify(args, H, modules='numpy')
__F = lambda X: _F(*(X.tolist()[0]))
__G = lambda X: _G(*(X.tolist()[0]))
__J = lambda X: _J(*(X.tolist()[0]))
__H = lambda X: _H(*(X.tolist()[0]))
```

#### Iteration

A couple of notes:

* The step size is formulated above as $\alpha_{n+1} = \text{argmax }_{\alpha \in \mathbb{R}} G(x_n + \alpha v_n)~$ which is a maximization problem, and since we're using scipy.optimize.minimize we minimize $-G(x_n + \alpha v_n)$.
* Recall that $u'(c_t) = c^{-\sigma}$. Here, $\sigma = 1.01$, and so if $c \lt 0~$, we're raising a negative number to a fractional power, which results in a complex number. Python will give us an error if we try to do this without indicating that we're okay with complex numbers, so in the minimization routine we add $0j$ to our vector, and then take only the real part of alpha (I don't know if this is okay...)

```python
import timeit

# Specify tolerance
Tmax = 1000;
eh   = 0.000000001;
es   = 0.000000001;

def iterate():
    # Guess solution
    X = [[0.1+0j]*(T+1) + [3+0j]*(T-1)]
    v = -np.dot(_J(*X[-1]), np.linalg.inv(_H(*X[-1])))
    step = minimize(lambda alpha: real(-__G(X[-1] + alpha*v)), 0)['x']
    X.append(X[-1] + step*v)
    
    while True:
        X.append(X[-1] + step*v)
        v = -np.dot(__J(X[-1]), np.linalg.inv(__H(X[-1])))
        step = minimize(lambda alpha: real(-__G(X[-1] + alpha*v)), 0)['x']
        
        # Break out of the iteration if we are sufficiently close to the zero
        if np.linalg.norm(__F(X[-1])) < es:
            break;
        # Break out of the iteration if we aren't moving anymore
        if np.linalg.norm(X[-1] - X[-2])/(1 + np.linalg.norm(X[-1])) < eh:
            #print 'Movement has stopped!';
            break;
        # Break out of the iteration if we've exceeded max
        if len(X) > Tmax-1:
            print 'Stopping criteria not met!';
            break;
            
    return X

# Run it once to get the coordinates
X = iterate()

# Run it again to get the timing
time = timeit.timeit('iterate()', 'from __main__ import iterate', number=1)
print "Iteration required %d iterations and %f seconds" % (len(X), time)

# Extract the consumption and capital vectors from the total resultant vector
_X = real(X[-1].A[0])
c = _X[:T+1]
# note that k0 and kT were given, so were not in the resultant vector
k = np.hstack(([k0], _X[T+1:], [kT]))
```

```python
import matplotlib.pyplot as plt

# Plot the results
fig   = plt.figure(figsize=(14,4));

paths = [(c,css,"Consumption"),(k,kss,"Capital")]
for i in range(len(paths)):
    path,ss,name = paths[i]
    
    # Create the new axes
    axes  = fig.add_subplot(1,len(paths),i+1);

    # plot path
    line, = axes.plot(range(T+1), path, color='#aaaaaa', linewidth=0.5); # note the tuple unpacking here

    # add a stem plot
    marker, stems, base = axes.stem(range(T+1), path, ':', bottom=ss);

    # and a line showing equilibrium capital
    axes.plot(range(T+1), [ss]*(T+1), color='gray');

    # prettify axes and lables
    axes.set_title(r'Ramsey: %s Path' % name);
    axes.set_xlabel('Time');
    axes.set_ylabel(name);
    axes.set_ylim(min(path)-0.2, max(path)+0.2);

    # prettify markers and stems
    map(lambda stem: stem.set(color='gray'), stems);
    marker.set(color='#999999', markersize=3);
```

```python
# Get the differential equations defining steady state
__kss = lambdify((kt, ct), (f(kt) - d*kt - ct), modules='numpy')
__css = lambdify((kt, ct), (1/s)*(f(kt).diff() - (1-b) - d)*ct, modules='numpy')

# Get a coordinate mesh for plotting the contour lines
K,C = np.meshgrid(
    np.linspace(0, 100, 100),
    np.linspace(0, 4, 100)
);

# Create the figure
fig = plt.figure(figsize=(10,10));
axes = fig.add_subplot(1,1,1);

# Plot off-steady-state streamlines
axes.streamplot(K+0.1, C+0.1, __kss(K+0.1,C+0.1), __css(K+0.1,C+0.1), color='#bbbbbb')

# Plot the steady state contours
axes.contour(K, C, __kss(K,C), 0, linewidths=1, colors='#000000'); # kss
axes.vlines(kss, 0, 4, colors='#000000')

# Plot saddle path
axes.plot(k, c, color='cornflowerblue', linewidth=2)

# Remove the top and right spines
axes.spines['top'].set(visible=False)
axes.spines['right'].set(visible=False)

# Remove tick marks and labels
axes.tick_params(length=0)

axes.set_title('Ramsey Model: Phase Diagram');
axes.set_xlabel('Capital');
axes.set_ylabel('Consumption');
```
