```python
import numpy as np
import scipy.linalg

# Model parameters
a = 0.5;  # capital's share of income
b = 0.95; # discount rate
d = 0.1;  # depreciation rate
s = 1.01; # coefficient of relative risk aversion

# Steady-state values
kss = (((1/b) + d - 1)/a)**(1/(a-1));
css = kss**a - d*kss;

# Functional specifications
def u(c):
  return (1 / (1 - s)) * (c**(1 - s) - 1);
def du(c):
  return c**(- s) - (1 / (1 - s));
def f(k):
  return k**a;
def df(k):
  return a*(k**(a-1));

# Linearization
phi = b*a*(1-a)*(kss**a)/s;
B = np.matrix([[1,phi],[0,1]]);
A = np.matrix([[1,0],[(-css/kss),1 + a*(kss**(a-1)) - d]]);
M = np.linalg.inv(B)*A;

# Eigenvalue decomposition
vals, vecs = np.linalg.eig(M);
S = np.linalg.inv(vecs);
Lambda = np.diag(vals);

# Generalize Schur decomposition (sort not working)
#S2, T, Q, Z = scipy.linalg.qz(B, A, sort='ouc');
#ree_schur_coef = -Z[1,1] / Z[1,0];

# Is it saddle path stable (determinate), with largest eig first?
test = np.linalg.norm(M - np.linalg.inv(S)*Lambda*S);
if not (test < 10e-8 and vals[0] > 1 and 1 > vals[1] and vals[0] > vals[1]):
  display("Did not work.");
reecoef = -S[0,1] / S[0,0];

#display(reecoef)
#display(ree_schur_coef)

# Time path in two ways
k0 = 0.6;
c0 = reecoef*k0;
T = 120;

M1 = (-css/kss)*reecoef + 1 + a*(kss**(a-1)) - d;

# Iterate to get paths under each method
# under method 1, where we take powers of the inverse matrix times the initial vector
# to get the time paths of both consumption and capital.
# note that this is vulnerable to computational rounding error problems
x = [np.matrix([c0, k0]).getT()];
# under method 2, where we take the linearized solution for ct (in terms of kt) and
# plug it directly into the capital accumulation equation, then we can simply iterate
# that equation to get the path of capital.
k_alt = [k0];

# Set iterations
T = 120;

# Get capital paths
for i in range(T-1):
  x.append(
    np.dot(M,x[-1])
  );
  k_alt.append(
    k_alt[-1]*M1 # ???
  );
    
# Extract the capital path for the first method
k = [t[1,0] for t in x];

# Plot the results
fig   = plt.figure();
axes  = fig.add_subplot(1,1,1);

# and a line showing equilibrium capital
axes.plot(range(T), [0]*T, color='gray');

# plot capital path from method 1
line, = axes.plot(range(T), k, color='red'); # note the tuple unpacking here
# plot capital path from method 2
line, = axes.plot(range(T), k_alt, color='blue', linestyle='dashed', linewidth=2); # note the tuple unpacking here

# add a stem plot

# prettify axes and lables
axes.set_title('Ramsey Capital Path');
axes.set_xlabel('Time');
axes.set_ylabel('Capital');
axes.set_ylim(-0.2, 1);

# Show the plot
#display(fig);
```
