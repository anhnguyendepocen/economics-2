__Model setup__

```python
import numpy as np
import sympy as sp
from sympy import Derivative as D
import math
import matplotlib.pyplot as plt

# Model variables
c,k = sp.symbols('c k', real=True);

# Model (abstract) parameters
a,b,d,s = sp.symbols('alpha beta delta sigma');

# Model (calibrated) parameters
calibration = {
    a: 0.5,  # capital's share of income
    b: 0.95, # discount rate
    d: 0.1,  # depreciation rate
    s: 1.01, # coefficient of relative risk aversion
}

# Model (abstract) functions
u = sp.Function('u'); # utility function
f = sp.Function('f'); # production function

# Model (explicit) functional forms
forms = {
    u(c): (c**(1-s) - 1) / (1 - s),
    f(k): k**a
}

# Get steady state values
kss = (((1/b + d - 1)/a)**(1/(a - 1))).xreplace(calibration).doit();
css = (kss**a - d*kss).xreplace(calibration).doit();

# Add steady state consumption to calibration
calibration[c] = css;
calibration[k] = kss;

# IFT functions
H = sp.Function('H');
h_ = sp.Function('h');
G = D(u(h_(k)), h_(k)) - b * (1 + D(f(H(k)), H(k)) - d) * D(u(h_(H(k))), h_(H(k)))
```

__Define h__

```python
# Objective function
class h(sp.Function):
    """
    Notes:
    all series have k in terms of deviation from steady-state

    h(k).series(n=4) uses nseries, returns with O() notation
    h(k).series() uses lseries, returns
    """
    nargs=1
    _DG = []
    
    @classmethod
    def DG(cls, k, n, rules):
        i = n-1 # the index of the derivative; we don't care about the 0th derivative
        if not len(cls._DG) > i:
            DG = D(G, k, n).doit();
        
            # Perform rule replacement until nothing changes
            while True:
                tmp = DG.xreplace(dict(rules)).doit()
                if sp.Eq(tmp, DG): break;
                DG = tmp
            
            cls._DG.append(DG.xreplace(calibration).doit())
        return cls._DG[i]
    
    @classmethod
    @cacheit
    def diff(cls, k, n):
        """Returns D^n(h(k)) evaluated at k = kss"""
        x = sp.symbols('x:'+`n+1`);
        
        # Get the lower order derivatives
        _Dh = []
        for i in range(n)[::-1]:
            _Dh.append(cls.diff(k, i))
        _Dh.reverse()
        
        # n == 0 is easy
        if n == 0: return css
        
        # Generate replacement rules
        rules = []
        for j in range(n,0,-1):
            rules += [
                (H(k).diff(k,j), D(f(k) + (1-d)*k -h_(k), k, j)), # H(k) -> (...)
                (h_(k).diff(k,j), x[j])                           # h^n(k) -> x_n
            ]
        rules += [(H(k), k), (h_(k), c)]  # Steady state
        rules += forms.items()            # Functional forms
    
        # Get D^n(G) evaluated at the steady state
        DG = cls.DG(k,n,rules).xreplace(dict(zip(x[:-1], _Dh)))
        # Solve for D^n(h)
        Dh = sp.solve(DG, x[n], simplify=False)
        
        # if n == 1, we know we have a quadratic, need to 
        # select the root that puts us on the saddle path
        if n == 1:
            DH0_kss = (forms[f(k)].diff(k) + (1-d) - Dh[0]).xreplace(calibration);
            Dh[0] = Dh[0] if abs(DH0_kss) < 1 else Dh[1]
        
        # Last, substitute in the lower-order derivatives
        return Dh[0]
```

__Plot the approximations of $h()$__

```python
# Parameters
n = 4;

# Get the series as an iterator
h_approx = h(k).series(n=None)

# Plot up to n approximations
fig = plt.figure(figsize=(16,8));
ax = fig.add_subplot(111);
k_coords = np.linspace(-100,100,100);

i = 0;
_c_coords = h_approx.next()
ax.hlines(_c_coords,-100,100)
for _h in h_approx:
    __h = lambdify(k, _h.removeO(), modules='numpy');
    _c_coords += __h(k_coords)
    ax.plot(k_coords, _c_coords)
    
    i += 1;
    if(i > n):
        break
ax.set(ylim=(-500,500));
```

__Plot the path of capital__

```python
# Parameters
n = 4;
T = 50;
k0 = kss - 5;

# Get the series as an iterator
h_approx = h(k).series(n=None)

# Get H(k) (except the h(k) term)
k_approx = (forms[f(k)](k) + (1-d)*k).xreplace({a:calibration[a], d:calibration[d]})
_k = lambdify(k, k_approx, modules='numpy');

# Plot up to n approximations
fig = plt.figure(figsize=(10,10));
ax = fig.add_subplot(111);
ax.hlines(kss,0,T-1, linestyles='dashed')

t_coords = range(T)
__h = h_approx.next()
i = 0
for _h in h_approx:
    __h += _h
    ___h = lambdify(k, __h, modules='numpy');
    
    k_coords = [k0]
    for t in t_coords:
        k_coords.append(_k(k_coords[-1]) - ___h(k_coords[-1]-kss));
    
    ax.plot(t_coords, k_coords[1:])
    i += 1;
    if(i > n):
        break
```
