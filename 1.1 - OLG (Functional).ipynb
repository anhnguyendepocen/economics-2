### 1.1 - OLG (Functional)

```python
"""OLG example code: Python in a functional style"""
import matplotlib.pyplot as plt

# Define functions
def get_kss(alpha):
  return ((1 - alpha) / 2)**((1 / (1 - alpha)));

def generate_forward(alpha):
  def forward(k):
    return ((1 - alpha) / 2) * (k**alpha);
  return forward;

def generate_simulate(alpha, Tmax, epsilon, delta):
  forward = generate_forward(alpha);
  kss = get_kss(alpha);
  def simulate(k):
    k = k + (forward(k[-1]),);

    if abs(k[-1] - kss) < epsilon or abs(k[-1] - k[-2]) < delta:
      return k;
    if len(k) > Tmax-1:
      print 'Stopping criteria not met!';
      return k;

    return simulate(k); # tail call recursion, not proper
  return simulate;

# Generate the simulation function
simulate  = generate_simulate(0.3, 100, 0.001, 0.001);

# Run the simulation
k = simulate((0.01,));

# Plot the results
alpha = 0.3;
kss = get_kss(alpha);
T = len(k);
fig   = plt.figure();
axes  = fig.add_subplot(1,1,1);

# plot capital path
line, = axes.plot(range(T), k, color='black'); # note the tuple unpacking here

# add a stem plot
marker, stems, base = axes.stem(range(T), k, ':');

# and a line showing equilibrium capital
axes.plot(range(T), [kss]*T, color='gray');

# prettify axes and lables
axes.set_title(r'OLG Capital Path, $\alpha = %s$' % alpha);
axes.set_xlabel('Time');
axes.set_ylabel('Capital');
axes.set_ylim(0, kss+0.2);

# prettify colors
map(lambda stem: stem.set(color='gray'), stems);
marker.set(color='gray');
```
